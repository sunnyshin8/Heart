```python
# specifying input directory
import pathlib 
Normal_o2s= pathlib.Path("data/lt98.csv") # o2 saturation less than 98
little_high=pathlib.Path("data/gt98.csv") # o2 saturation greater than 98
```

```python
# converting csv files to list
import csv
results = {} # dictionary to store data
dirlist = [Normal_o2s, little_high] # list of files
classes = ['No', 'Yes'] # list of classes
for file, list in zip(classes, dirlist): # looping through files and classes
    with open(list) as csvfile: # opening file
        reader = csv.reader(csvfile) # reading file
        count = 0 # initializing count
        for row in reader: # looping through rows
            count += 1 # incrementing count
        results[list] = count # storing count in dictionary

for key, value in results.items(): # looping through results
    print(key, value) # printing key and value where key is file and value is count
```

```python
from sklearn.preprocessing import StandardScaler # importing StandardScaler
X = df.drop('output', axis=1) # dropping output column to create X
y = df['output'] # only output column to create y
scaler = StandardScaler() # creating an instance of StandardScaler
scaler.fit(X) # fitting the data
X_scaled = scaler.transform(X) # transforming the data
X_scaled = pd.DataFrame(X_scaled, columns=X.columns) # converting to dataframe


display(X.head()) # displaying first 5 rows of X
display(y.head()) # displaying first 5 rows of y
```

```python
saturation = pd.read_csv('data/o2Saturation.csv') # reading csv file
saturation = saturation.rename(columns={'98.6' : 'o2Saturation'}) # renaming column
saturation.head() # displaying first 5 rows
saturation.shape # displaying shape of dataframe
```

```python
# Train test split
# importing train_test_split
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold, RandomizedSearchCV, cross_val_score
# stratify=y to maintain the ratio of classes in train and test data
raw_train_X, raw_test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, stratify=y)

scaler = StandardScaler() # creating an instance of StandardScaler
scaler.fit(raw_train_X) # fitting the data
train_X_scaled = scaler.transform(raw_train_X) # transforming train data
test_X_scaled = scaler.transform(raw_test_X) # transforming test data
```

```python
# Decision Tree Classifier
from scipy.stats import randint # importing randint
from sklearn.tree import DecisionTreeClassifier # importing DecisionTreeClassifier
paramns = { # creating a dictionary of parameters
    'criterion' : ['gini', 'entropy'],
    "max_depth" : [1, 3, 6, 9, 18, 24, None],
    "min_samples_split" : randint(32, 128),
    "min_samples_leaf" : randint(32, 128),
}

# Randomized Search CV for Decision Tree
search_tree = RandomizedSearchCV(DecisionTreeClassifier(), 
                    paramns,
                    cv = StratifiedKFold(n_splits = 5, shuffle=True), n_iter=62)
search_tree.fit(train_X_scaled, train_y) # fitting the data
results_tree = pd.DataFrame(search_tree.cv_results_) # converting results to dataframe
results_tree.head() # displaying first 5 rows
```

```python
results_tree.shape # displaying shape of dataframe

# Note: 
# The first entry is the number of iterations (n_iter) 
# The second entry is the number of coulumns
```

```python
# cross validation score
scores_tree = cross_val_score(search_tree, X, y, cv=StratifiedKFold(n_splits=6, shuffle=True))
scores_tree
```

```python
search_tree.best_estimator_ # best estimator
```

```python
search_tree.best_score_ # best score
```

```python
# Random Forest Classifier
# importing RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

# Randomized Search CV for Random Forest
search_random_forest = RandomizedSearchCV(RandomForestClassifier(),
                    paramns,
                    cv = StratifiedKFold(n_splits = 5, shuffle=True), n_iter=62)
search_random_forest.fit(train_X_scaled, train_y) # fitting the data
results_random_forest = pd.DataFrame(search_random_forest.cv_results_) # converting results to dataframe
results_random_forest.head() # displaying first 5 rows
```

```python
results_random_forest.shape # displaying shape of dataframe

# Note:
# The first entry is the number of iterations (n_iter)
# The second entry is the number of columns
```

```python
# Support Vector Classifier
from sklearn.svm import SVC # importing SVC
model_svc = SVC() # creating an instance of SVC
model_svc.fit(X, y) # fitting the data
# cross validation score
cross_svc = cross_validate(model_svc, X, y, cv=StratifiedKFold(n_splits=11, shuffle=True))

mean_svc = cross_svc['test_score'].mean() # mean of test score
std_svc = cross_svc['test_score'].std() # standard deviation of test score

print(mean_svc * 100) 
print('Accuracy: [%.2f, %.2f]' % ((mean_svc - 1.99 * std_svc) * 100, (mean_svc + 1.99 * std_svc) * 100))
```

```python
# Bernoulli Naive Bayes
from sklearn.naive_bayes import BernoulliNB # importing BernoulliNB
model_ber = BernoulliNB() # creating an instance of BernoulliNB
model_ber.fit(train_X_scaled, train_y) # fitting the data
# cross validation score
cross_ber = cross_validate(model_ber, X, y, cv=StratifiedKFold(n_splits=7, shuffle=True))

mean_ber = cross_ber['test_score'].mean() # mean of test score
std_ber = cross_ber['test_score'].std() # standard deviation of test score

print(mean_ber * 100)
print('Accuracy: [%.2f, %.2f]' % ((mean_ber - 1.99 * std_ber) * 100, (mean_ber + 1.99 * std_ber) * 100))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# function to get the metrics
def get_metrics(test_y, predict):
    accuracy = accuracy_score(test_y, predict) * 100
    precision = precision_score(test_y, predict) * 100
    recall = recall_score(test_y, predict) * 100
    f1 = f1_score(test_y, predict) * 100
    return accuracy, precision, recall, f1
```

```python
# Getting the best estimator for random forest
model_rfc = search_random_forest.best_estimator_
model_rfc.fit(train_X_scaled, train_y)
predict_rfc = model_rfc.predict(test_X_scaled)
# getting the metrics for random forest
accuracy_rfc, precision_rfc, recall_rfc, f1_rfc = get_metrics(test_y, predict_rfc)
```

```python

# bernoulli naive bayes model
model_ber = BernoulliNB()
model_ber.fit(train_X_scaled, train_y)
predict_ber = model_ber.predict(test_X_scaled)
# getting the metrics for bernoulli naive bayes
accuracy_ber, precision_ber, recall_ber, f1_ber = get_metrics(test_y, predict_ber)
```

```python
# Indexes to be used in dataframe
index = ['Accuracy', 'Precision', 'Recall', 'F1']
```

```python
# Creating a dataframe with random forest metrics
compare = pd.DataFrame([accuracy_rfc, precision_rfc, recall_rfc, f1_rfc], columns=['RandomForestClassifier'], index=index)
# Adding bernoulli naive bayes metrics to the dataframe
compare = compare.assign(Naive = [accuracy_ber, precision_ber, recall_ber, f1_ber])
# rounding the values to 2 decimal places
compare = compare.round(2)
compare
```
